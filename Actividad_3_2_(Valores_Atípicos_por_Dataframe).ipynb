# Actividad 3.2 (Valores Atípicos por Dataframe)

```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar archivo csv desde equipo
from google.colab import files
files.upload()
```

```python
data2020= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2020')
data2021= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2021')
data2022= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2022')
data2023= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2023')
```

## 2020

```python
valores_nulos2020=data2020.isnull().sum()
valores_nulos2020
```

```python
## Remplazar valores nulos 
data2020SN =data2020.fillna(method="bfill") 
data2020SN=data2020SN.drop(['POLIZA'], axis=1)
data2020SN
```

```python
valores_nulos2020=data2020SN.isnull().sum()
valores_nulos2020
```

```python
## Eliminar outliers con desviación estándar
fig = plt.figure(figsize =(15, 8))
data2020.plot(kind='box', vert=False)
plt.title("Valores Atípicos 2020")
plt.show() #dibujamos el diagrama antes de manipulación
```

```python
#Eliminamos columnas de tipo string
data2020num= data2020.drop(['FECHA','UUID','FOLIO','RFC','PROVEEDOR','TIPO GASTO','GASTO','DESCRIPCION','TIPO','STATUS','Columna1','POLIZA'], axis=1)
data2020num
```

```python
y=data2020num
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
outliers= data2020num[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
data_clean2020= data2020num[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean2020
```

```python
data_clean2020std=data_clean2020.copy()
data_clean2020std=data_clean2020std.fillna(method='bfill')
data_clean2020std
```

```python
valores_nulos=data_clean2020std.isnull().sum()
valores_nulos
```

```python
fig = plt.figure(figsize =(15, 8))
data_clean2020std.plot(kind='box', vert=False)
plt.title("Valores de 2020")
plt.show() #dibujamos el diagrama
```

```python
## Remplazar valores outliers con cuantiles
y=data2020num

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
outliers_iqr= data2020num[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data_clean2020_iqr= data2020num[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean2020_iqr
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data2020iqr=data_clean2020_iqr.copy()
data2020iqr=data2020iqr.fillna(method='bfill')
data2020iqr
```

```python
fecha=data2020['FOLIO']
data2020iqr = pd.concat([fecha, data2020iqr], axis=1)
data2020iqr
```

```python
fig = plt.figure(figsize =(15, 8))
data2020iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos 2020")
plt.show() #dibujamos el diagrama
```
